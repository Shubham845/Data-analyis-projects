# Eliminating NULL Values

```python

#We can drop these null values because these are only 0.01% of the dataset.
df=df.dropna(axis=0,subset=['rating'])
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=0,subset=['writer','released','score','votes','country','runtime','company'])
```

```python
df.budget.fillna(df.budget.mean(),inplace=True)
```

```python
df.gross.fillna(df.gross.mean(),inplace=True)
```

```python
df.isnull().sum()
```

# Checking and changing Datatypes

```python
df.dtypes
```

```python
#change data types of columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
#creating correct date column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df.head()
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
sns.regplot(x="budget",y="gross",data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})


plt.show()
```

```python
df.corr()
```

```python
correlation_matrix=df.corr()

sns.heatmap(correlation_matrix,annot=True)
plt.title('corr_numeric_features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized=df.copy()

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
    
df_numerized.head()
```
